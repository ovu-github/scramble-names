```python
import os
import pandas as pd
import random
folder = '/media/m/1083-4BEC/'
```

```python
mp3s = []

[mp3s.append(file) for file in os.listdir(folder) if '.mp3' in file]


mp3s
```

```python
random.shuffle(mp3s)
mp3s
```

```python
nl = '\n'

for count, file in enumerate(mp3s):
    print(f'will rename: {folder + file} to {nl} {folder + str(count) + file}')
    os.rename(folder + file,
              folder + str(count) + file)
    
```
